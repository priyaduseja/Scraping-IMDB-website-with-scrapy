##### Extracting Movie Name, Duration and Genres of the movie from IMDB website

```python
import scrapy
from scrapy.crawler import CrawlerProcess

class IMDB_Movie(scrapy.Spider):
    name = 'Movies'
    start_urls = ['https://www.imdb.com/chart/top/']
    custom_settings = {
        'DEFAULT_REQUEST_HEADERS': {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36',
        }
    }
    
    def parse(self, response):
        for movie in response.css('ul.ipc-metadata-list.ipc-metadata-list--dividers-between.sc-a1e81754-0.eBRbsI.compact-list-view.ipc-metadata-list--base li div.ipc-metadata-list-summary-item__c a'):
            movieName = movie.css('::text').get()
            movieUrl = movie.css('::attr(href)').get()
            dic = {'movieName':movieName }
            yield response.follow(movieUrl, callback= self.parseMovieInfo, meta=dic)
            
    def parseMovieInfo(self, response):
        movieName = response.css('div.sc-67fa2588-0.cFndlt h1 span::text').get()
        movieNameParse = response.meta['movieName']
        movieDuration = response.css('ul.ipc-inline-list.ipc-inline-list--show-dividers.sc-d8941411-2.cdJsTz.baseAlt li:last-child::text').get()
        genres = response.css('.ipc-chip-list__scroller a span::text').getall()
        genres = ','.join(genres)
        yield {'Movie Name from parseMovieInfo' : movieNameParse, 'Movie Duration' : movieDuration, 'Movie Genres': genres}
        

process = CrawlerProcess(settings = {'FEEDS': {'IMDB_Movies_Duration.csv': {'format': 'csv'}}})
process.crawl(IMDB_Movie)
process.start()
```

#### Extracting the movie name and year from the first page and the complete release data from the leading page. The data is then exported to a csv file

```python
import scrapy
from scrapy.crawler import CrawlerProcess

class Movies(scrapy.Spider):
    name = 'Movies'
    start_urls = ['https://www.imdb.com/chart/top/']
    custom_settings = {'DEFAULT_REQUEST_HEADERS': {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36',}}
    
    def parse(self, response):
        for movie in response.css('ul.ipc-metadata-list li'):
            movieName = movie.css('div.ipc-title a h3::text').get()
            movieYear = movie.css('div.sc-b0691f29-7 span:first-child::text').get()
            movieUrl = movie.css('div.ipc-title a::attr(href)').get()
            dic = {'MovieName': movieName, 'MovieYear': movieYear}
            yield response.follow(movieUrl, callback=self.parseMovieInfo, meta = dic)
    
    def parseMovieInfo(self, response):
        movieNameParse = response.meta['MovieName']
        movieYearParse = response.meta['MovieYear']
        movieReleaseDate = response.css('li[data-testid="title-details-releasedate"] a.ipc-metadata-list-item__list-content-item::text').get()
        yield {'MovieName': movieNameParse, 'MovieYear' : movieYearParse, 'MovieReleaseDate': movieReleaseDate }
            
process = CrawlerProcess(settings = {'FEEDS': {'IMDB_Movies_Year_ReleaseDate.csv': {'format': 'csv'}}})
process.crawl(Movies)
process.start()
```
